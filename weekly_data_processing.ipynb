```python
import pandas as pd
import numpy as np

weekly = pd.read_csv("weekly_data.csv")
```

```python
weekly.head()
```

```python
weekly_last = weekly.tail(52).reset_index()
```

```python
wl = weekly_last.reset_index()
wl["level_0"] = wl["level_0"]+1
wll = wl[["level_0", "revenue"]]
wll.rename(columns={"level_0": "Week"}, inplace=True) 
wll.rename(columns={"revenue": "Revenue"}, inplace=True) 
```

```python
wll.to_csv("wll.csv", index=True)
```
